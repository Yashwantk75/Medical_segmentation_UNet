# We can see that image and mask are not same in number that means some image have more than one mask so we need to combine them

0 to 436 we have Benign class,
437 to 646 we have Malignent class and
647 to 780 we have Normal class

```python
import cv2
```

```python
def load_image(path,size):
    image=cv2.imread(path)
    image=cv2.resize(image,(size,size))
    image=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
    image=image/255
    return image
```

```python
def load_data(paths,size):
    images=[]
    masks=[]

    x=0

    for path in sorted(paths):
        img = load_image(path,size)

        if 'mask' in path:
            if x:
                masks[-1]+=img
                masks[-1]=np.array(masks[-1]>0.5,dtype='float64')
            else:
                masks.append(img)
                x=1
        else:
            images.append(img)
            x=0
    return np.array(images),np.array(masks)
```

```python
size=128
x,y=load_data(paths,size)
```

## Modeling

### Building U-net architecture

#### Convo Block

```python
def conv_block(input,num_of_filters):
    conv=tf.keras.layers.Conv2D(num_of_filters,(3,3),activation='relu',padding='same',kernel_initializer='he_normal')(input)
    conv=tf.keras.layers.Conv2D(num_of_filters,(3,3),activation='relu',padding='same',kernel_initializer='he_normal')(conv)
    return conv
```

#### Encoder Block

```python
def encoder_block(input,num_of_filters):
    conv=conv_block(input,num_of_filters)
    pool=tf.keras.layers.MaxPool2D((2,2))(conv)
    return conv,pool
```

#### Decoder block

```python
def decoder_block(input, skip_features,num_of_filters):
    upconv=tf.keras.layers.Conv2DTranspose(num_of_filters,(2,2),strides=2,padding='same')(input)
    con=tf.keras.layers.concatenate([upconv,skip_features])
    conv=conv_block(con,num_of_filters)
    return conv
```

### Build Model

```python
def build_model(input_shape):
    input_layer=tf.keras.layers.Input(input_shape)

    s1,p1=encoder_block(input_layer,64)
    s2,p2=encoder_block(p1,128)
    s3,p3=encoder_block(p2,256)
    s4,p4=encoder_block(p3,512)

    b1 = conv_block(p4,1024)

    d1=decoder_block(b1,s4,512)
    d2=decoder_block(d1,s3,256)
    d3=decoder_block(d2,s2,128)
    d4=decoder_block(d3,s1,64)

    output_layer=tf.keras.layers.Conv2D(1,1,padding='same',activation='sigmoid')(d4)
    model=tf.keras.Model(input_layer,output_layer,name='U-Net')

    return model
```

```python
model=build_model(input_shape=(size,size,1))
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.summary()
```

# Training

```python
y_pred=model.predict(X_test,verbose=0)
y_pred_thresholded = y_pred > 0.5

IOU_keras = MeanIoU(num_classes=2)
IOU_keras.update_state(y_pred_thresholded, y_test)
print("Mean IoU =", IOU_keras.result().numpy())

prec_score = Precision()
prec_score.update_state(y_pred_thresholded, y_test)
p = prec_score.result().numpy()
print('Precision Score = %.3f' % p)

recall_score = Recall()
recall_score.update_state(y_pred_thresholded, y_test)
r = recall_score.result().numpy()
print('Recall Score = %.3f' % r)

f1_score = 2*(p*r)/(p+r)
print('F1 Score = %.3f' % f1_score)
```
